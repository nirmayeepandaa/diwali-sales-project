# EXPLORATORY DATA ANALYSIS

## Gender

```python
df.columns
```

```python
custom_palette ={'F':'darkblue','M': 'lightblue'}
ax=sns.countplot(x='Gender',data=df,palette = custom_palette,hue='Gender',legend=False)
sns.set_theme(rc={'figure.figsize':(5,2)})
for bars in ax.containers:
    ax.bar_label(bars)
    ax.set_xticks(ax.get_xticks())
    ax.set_xticklabels(['Female','Male'])

plt.title('Distribution of Gender')
```

```python
custom_palette ={'F':'darkblue','M': 'lightblue'}
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax=sns.barplot(x='Gender',y='Amount',data=sales_gen,palette=custom_palette,hue='Gender',legend=False)

for bars in ax.containers:
    ax.set_xticks(ax.get_xticks())
    ax.set_xticklabels(['Female','Male'])
def format_inr_full(x,pos):
    return f'Rs.{int(x):,}'
ax.yaxis.set_major_formatter(format_inr_full)

for container in ax.containers:
    ax.bar_label(container,labels=[f'Rs.{int(bar.get_height()):,}' for bar in container])
```

*From above graphs we can see that most of the buyers are females and even the purchasing power power of females are greater than men.*

## Age

```python
custom_palette ={'F':'darkblue','M': 'lightblue'}
df['Age Group'] = pd.Categorical(df['Age Group'], categories=['0-17','18-25','26-35','36-45','46-50',"51-55",'55+'])
ax = sns.countplot(data=df,x='Age Group', hue = 'Gender',palette= custom_palette)

for bars in ax.containers:
    ax.bar_label(bars)
   
```

```python

def format_inr_full(x,pos):
    return f'Rs.{int(x):,}'
ax.yaxis.set_major_formatter(format_inr_full)
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount')
ax = sns.barplot(x='Age Group',y='Amount',data=sales_age,palette=['blue','green','yellow','red','orange','violet','blue'],hue='Age Group',legend=False)
for container in ax.containers:
    ax.bar_label(container,labels=[f'Rs.{int(bar.get_height()):,}' for bar in container])
sns.set_theme(rc={'figure.figsize':(11,7)})
```

*From the above graphs we can see that most of the buyers are of age group between 26-35 yrs female.*

## State

```python
#total number of order from top states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
ax = sns.set_theme(rc={'figure.figsize':(17,7)})
ax = sns.barplot(data = sales_state, x ='State',y = 'Orders',palette = sns.color_palette('Set2'),hue='State',legend=False)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
ax = sns.set_theme(rc={'figure.figsize':(17,7)})
ax = sns.barplot(data= sales_state,x='State',y='Amount',palette=sns.color_palette('Set1'),hue='State',legend=False)
def format_inr_full(x,pos):
    return f'Rs.{int(x):,}'
ax.yaxis.set_major_formatter(format_inr_full)
for container in ax.containers:
    ax.bar_label(container,labels=[f'Rs.{int(bar.get_height()):,}' for bar in container])
```

*From above graphs we can see that most of the orders & sales are from Uttar Pradesh,Maharashtra and Karnataka respectively.*
